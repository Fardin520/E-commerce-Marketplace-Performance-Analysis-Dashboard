# 01. Data Exploration - Olist E-commerce Dataset

## Objective
In this notebook, we will explore the structure of the Olist e-commerce dataset. We will understand each table, its columns, and how different tables are connected.

```python
import pandas as pd
import os
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

```python
folder_path = r'D:\My Personal Projects\Ecommerce_Marketplace_Analysis'

customers = pd.read_csv(os.path.join(folder_path, 'olist_customers_dataset.csv'))
orders = pd.read_csv(os.path.join(folder_path, 'olist_orders_dataset.csv'))
order_items = pd.read_csv(os.path.join(folder_path, 'olist_order_items_dataset.csv'))
products = pd.read_csv(os.path.join(folder_path, 'olist_products_dataset.csv'))
sellers = pd.read_csv(os.path.join(folder_path, 'olist_sellers_dataset.csv'))
payments = pd.read_csv(os.path.join(folder_path, 'olist_order_payments_dataset.csv'))
reviews = pd.read_csv(os.path.join(folder_path, 'olist_order_reviews_dataset.csv'))

print("All datasets loaded successfully!")
```

```python
datasets = {
    'Customers': customers,
    'Orders': orders,
    'Order Items': order_items,
    'Products': products,
    'Sellers': sellers,
    'Payments': payments,
    'Reviews': reviews
}

for name, df in datasets.items():
    print(f"\n{'='*60}")
    print(f"Table: {name}")
    print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns")
    print(f"Columns: {list(df.columns)}")
```

```python
print("ORDERS TABLE:")
print(orders.head(3))
print("\nORDER ITEMS TABLE: ")
print(order_items.head(3))
print("\nCUSTOMERS TABLE: ")
print(customers.head(3))
print("\nPRODUCTS TABLE: ")
print(products.head(3))
```

```python
print("ORDERS INFO: ")
print(orders.info())
print("\nMissing Values in Orders: ")
print(orders.isnull().sum())
```

```python
print("Order Status Distribution:")

print(orders['order_status'].value_counts())
```

```python
print("Unique Orders in 'orders' table:        ", orders['order_id'].nunique())
print("Unique Orders in 'order_items' table:   ", order_items['order_id'].nunique())
print("Unique Customers in 'customers' table:  ", customers['customer_id'].nunique())
print("Unique Products in 'products' table:    ", products['product_id'].nunique())
```

```python
print("Average items per order:")
print(order_items.groupby('order_id')['order_item_id'].count().mean())
```

```python
delivered_orders = orders[orders['order_status'] == 'delivered'].copy()
print("Total Delivered Orders: ", delivered_orders.shape[0])
```

```python
order_data = delivered_orders.merge(
    order_items, 
    on='order_id', 
    how='inner'
)

print("Shape after merging with order_items:", order_data.shape)
```

```python
order_data = order_data.merge(
    customers, 
    on='customer_id', 
    how='left'
)

print("Shape after merging with customers:", order_data.shape)
```

```python
order_data = order_data.merge(
    products, 
    on='product_id', 
    how='left'
)

print("Final merged dataset shape: ", order_data.shape)
print("\nColumns in the merged dataset: ")
print(list(order_data.columns))
```

```python
date_columns = [
    'order_purchase_timestamp', 
    'order_approved_at', 
    'order_delivered_carrier_date', 
    'order_delivered_customer_date', 
    'order_estimated_delivery_date',
    'shipping_limit_date'
]

# Convert to datetime
for col in date_columns:
    order_data[col] = pd.to_datetime(order_data[col], errors='coerce')

print("Date columns converted to datetime format.")
```

```python
print("Missing values in merged dataset:\n")
print(order_data.isnull().sum())
```

```python
print("Missing values in merged dataset:\n")
print(order_data.isnull().sum())
```

```python
# Fill missing product category with 'Unknown'
order_data_clean = order_data.copy()

# Fill missing product category with 'Unknown'
order_data_clean['product_category_name'] = order_data_clean['product_category_name'].fillna('Unknown')

# Fill missing product dimensions with median
product_cols = ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']
for col in product_cols:
    order_data_clean[col] = order_data_clean[col].fillna(order_data_clean[col].median())

print("Missing values handled successfully!")
print("\nRemaining missing values:")
print(order_data_clean.isnull().sum())

# Fill missing product dimensions with median
product_cols = ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']
for col in product_cols:
    order_data_clean[col] = order_data_clean[col].fillna(order_data_clean[col].median())

print("Missing values handled successfully!")
print("\nRemaining missing values:")
print(order_data_clean.isnull().sum())
```
